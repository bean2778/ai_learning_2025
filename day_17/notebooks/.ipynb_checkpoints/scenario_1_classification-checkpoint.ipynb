```python
"""
PROBLEM FORMULATION: Customer Purchase Prediction

1. PROBLEM TYPE: Binary Classification
   - Target: 'purchased' (0 or 1)
   - Predicting whether a customer will make a purchase

2. CLASS BALANCE: Imbalanced (80/20 split)
   - Challenge: Can't use accuracy as primary metric
   - A model predicting all 0's would get 80% accuracy but be useless

3. RECOMMENDED METRICS:
   Primary: F1-Score
     - Balances precision and recall
     - Good for imbalanced classification
   
   Secondary: ROC-AUC
     - Measures overall discrimination ability
     - Less sensitive to class imbalance
   
   Monitor: Precision
     - If marketing budget is limited, avoid false positives
     - Don't waste money targeting people who won't buy

4. FEATURE ANALYSIS:
   - Income (0.30): Strongest predictor - wealthier customers buy more
   - Previous purchases (0.25): Loyalty effect - past buyers return
   - Time on site (0.16): Engagement - browsers may convert
   - Email opens (0.08): Weak signal but still useful
   - Age, customer_id: No predictive value

5. SPLITTING STRATEGY:
   - 60% train, 20% validation, 20% test
   - MUST use stratify=y to maintain 80/20 ratio in all splits
   - Random seed = 42 for reproducibility
"""
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['purchased', 'customer_id'], axis=1)
y = df['purchased']

X_temp, X_test, y_temp, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

# Second split: separate validation from train (20% of original = 25% of temp)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp,
    test_size=0.25,  # 0.25 * 0.8 = 0.2 of original
    stratify=y_temp,
    random_state=42
)

# Verify splits
print("Split sizes:")
print(f"  Train: {len(X_train)} ({len(X_train)/len(df)*100:.1f}%)")
print(f"  Val:   {len(X_val)} ({len(X_val)/len(df)*100:.1f}%)")
print(f"  Test:  {len(X_test)} ({len(X_test)/len(df)*100:.1f}%)")

print("\nClass distribution:")
print(f"  Train: {y_train.value_counts().to_dict()}")
print(f"  Val:   {y_val.value_counts().to_dict()}")
print(f"  Test:  {y_test.value_counts().to_dict()}")

print("\nClass balance preserved:")
print(f"  Train: {y_train.mean():.1%} positive")
print(f"  Val:   {y_val.mean():.1%} positive")
print(f"  Test:  {y_test.mean():.1%} positive")
```
